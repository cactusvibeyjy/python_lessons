```python
#더이상 필요없는 열 삭제 'Name', 'Ticket', 'SibSp', 'Parch', 'PassengerId'
train = df.drop(['Name', 'Ticket', 'SibSp', 'Parch', 'PassengerId'], axis =1, inplace = False)
test = test_df.drop(['Name', 'Ticket', 'SibSp', 'Parch', 'PassengerId'], axis =1, inplace = False)
test.head()
```

```python
#Training set
#모델의 학습에 사용되는 데이터
#TValidation set 
#모델 제작 과정 중, 학습된 모델의 성능을 측정하기 위한 데이터

#test set 모델의 최종 성능을 평가하기 위한 데이터
```

```python
#훈련용 데이터에서 데이터, label(target)분리
train_data= train.drop('Survived', axis =1) #학습용 데이터
target_data = train['Survived'] #학습용데이터의 label (target)
train_data.head()
```

```python
target_data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, Y_train, Y_valid = train_test_split(train_data, target_data)
print(train_data.shape, X_train.shape, X_valid.shape)
```

```python
#sklearn 의 Decision Tree Classifier를 사용하여 학습
from sklearn.tree import DecisionTreeClassifier

#Decision Tree 로 모델 생성
tree_dec = DecisionTreeClassifier()

#학습
tree_dec.fit(X_train, Y_train)
```

```python
print('training set accuracy : ', tree_dec.score(X_train, Y_train))
print('training set accuracy : ', tree_dec.score(X_valid, Y_valid))
```

```python
#만들어진 Decision Tree 로 test데이터로 판단
prediction = tree_dec.predict(test)
prediction
```

```python
submit = pd.DataFrame({
    'PassengerId' : test_df['PassengerId'],
    'Survived' : prediction
    
})
submit.to_csv('survived.csv', index = False)
```
