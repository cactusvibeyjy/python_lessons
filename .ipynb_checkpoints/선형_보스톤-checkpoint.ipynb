```python
#보스톤 주택가격 데이터에 머신러닝 기반의 회귀 분석 수행 (p.306)
from sklearn.datasets import load_boston
import pandas as pd

boston = load_boston()
boston.DESCR
```

```python
boston.data
```

```python
boston.feature_names
```

```python
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
boston_df.head()
```

```python
#crim = crime
#zn = 25000 피트 초과 거주 지역
#indus = 비상업 지역의 넓이 비율
#chas = 찰스강의 더미 변수 (1 = 강의 경계, 0 =경계 아님)
#nox = 일산화 질소 농도
#RM = 거주할 수 있는 방 개수
#AGE = 1940년 이전에 건축된 주택 비율
#dis = 5개 주요 고용센터까지 가중 거리
#rad = 고속도로 접근 용이도 
#tax = 10,000 달러당 재산세 비율
#ptratio = 지역의 교사와 학생 수 비율
#b = 지역의 흑인 거주 비율
#last = 하위 게층 비율
```

```python
boston_df.shape
```

```python
boston.target
```

```python
boston_df['PRICE'] = boston.target
boston_df.head()
```

```python
boston_df.shape
```

```python
boston_df.info()
```

```python
boston_df.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# x, y (data, label) 분할하기
Y= boston_df['PRICE']
X= boston_df.drop(['PRICE'], axis =1, inplace = False)
Y
```

```python
#훈련용 데이터와 평가용 데이터 분할 하기
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state = 156) # 랜덤값을 맞추면 일정한 난수를 발생
```

```python
#선형 회귀 분석 모델 생성
lr = LinearRegression()
#훈련
lr.fit(X_train, Y_train)
```

```python
#선형 회귀 분석 : 평가 데이터에 대한 예측 수행 ==> 예측 결과
Y_predict = lr.predict(X_test)
Y_predict 
```

```python
from sklearn.metrics import mean_squared_error, r2_score
#MSE = 평가지표 r2_score = R2(결정계수)
import numpy as np
#사이킷런에서 제곱 스퀘어루트를 제공하지 않기에 넘파이를 이용
```

```python
#실제값 =Y_test 예측값 = Y_predict
```

```python
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)
print('mse:', mse, 'rmse: ', rmse)
print('R^2:', r2_score(Y_test, Y_predict))
#variance score = R^2 = Best possible score is 1.0, lower values are worse.
```

```python
print('Y절편 값 : ', lr.intercept_)
print('회귀계수(기울기)', np.round(lr.coef_,1))
```

```python
coef = pd.Series(data=np.round(lr.coef_,2), index = X.columns)
coef.sort_values(ascending  =False)
```
