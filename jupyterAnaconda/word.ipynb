```python
import wordcloud
import matplotlib.pyplot as plt
```

```python
keywords = {'안녕':2, '하세요':1, '빅데이터': 15, '웹크롤링':3}
```

```python
wc = wordcloud.WordCloud(font_path = 'c:/Windows/fonts/malgun.ttf')
cloud = wc.generate_from_frequencies(keywords)
```

```python
figure = plt.figure()
plt.imshow(cloud)
plt.show()
figure.savefig('word.png')
```
